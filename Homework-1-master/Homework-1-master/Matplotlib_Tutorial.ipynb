Matplotlib is the primary library used for plotting figure/graphs in Python. You'll see it used in Homework-1, and likely in future homeworks and classes. Here's a brief overview of its common functions.

```python
# Let's use see plots within notebook.
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt # Key import.
```

### Simple Plot

The basic plot function takes in two lists, the first represents the X part of your coordinate pairs, and the second represents the Y part. So a single pair of coordinates is (X[i], Y[i]). pyplot plots these points and draws a line through them.

plt.plot can also infer the X values if you give it just one list of numbers (the inferred values will be equivalent to list(range(0, N)) ).

```python
def simple_plot():
    X = [1, 2, 3, 4]
    Y = [1, 2, 3, 4]
    plt.plot(X, Y) 
    plt.plot(Y) # Shifted to left because the inferred X[0] = 0.
    plt.ylabel('some numbers')
    plt.show()
    
simple_plot()
```

### Formatted Plot

Formatting the plot is easy. plt.plot has a third argument you can use to specify the color and type of plot you'd like. Notice the plot below is now black and discrete points rather than a continuous line. (Default value for this argument is 'b-', a blue line.)

```python
def formatted_plot():
    X = [1, 2, 3, 4]
    Y = [1, 2, 3, 4]
    plt.plot(X, Y, 'go') 
    plt.show()
    
formatted_plot()
```

### Common Plots

Three common matplotlib functions you'll use are bar, scatter, and plot. Notice for our X data, we're using strings instead of ints/floats. Matplotlib can deal with this.

```python
def common_plots():
    X = ['A', 'B', 'C']
    Y = [1, 10, 100]

    plt.bar(X, Y, color='g')
    plt.show()
    
    plt.scatter(X, Y, color='r')
    plt.show()
    
    plt.plot(X, Y, color='m')
    plt.show()
    
common_plots()
```

### Using Numpy

Matplotlib works great with numpy. You can pass numpy lists in place of plain Python lists and matplotlib will handle the input properly.

```python
def numpy_plot():
    t = np.arange(-3., 3., 0.2)
    plt.plot(t, t, 'ro')
    plt.plot(t, t**2, 'bo')
    plt.plot(t, t**3, 'go')
    plt.show()
    
numpy_plot()
```

### Using Dictionaries

You can also use dictionaries. Pass in the key to the argument slot, the dictionary itself into "data", and numpy will use the passed keys to look-up the relevant data within your dictionary.

```python
def keyword_plot():
    num = 30
    
    dictionary = {
        'x_vals': np.arange(num),
        'y_vals': np.arange(num) + 10 * np.random.randn(num),
        'color': np.random.randint(0, 10, num),
        'size': np.abs(np.random.randn(num)) * 100
    }
    
    plt.scatter('x_vals', 'y_vals', c='color', s='size', data=dictionary)
    plt.xlabel('x_vals')
    plt.ylabel('y_vals')
    plt.show()
    
keyword_plot()
```

### Sub Plots

You can also graph multiple plots in one figure using plt.subplot. From the documentation:

"The subplot() command specifies numrows, numcols, plot_number where plot_number ranges from 1 to numrows x numcols. The commas in the subplot command are optional if numrows x numcols<10. So subplot(211) is identical to subplot(2, 1, 1)."

```python
def sub_plots():
    t1 = np.arange(0.0, 5.0, 0.1)
    y1 = np.exp(-t1) * np.sin(2*np.pi*t1)
    t2 = np.arange(0.0, 5.0, 0.02)
    y2 = np.sin(2*np.pi*t2)

    plt.figure()
    plt.subplot(211)
    plt.plot(t1, y1, 'bo')

    plt.subplot(212)
    plt.plot(t2, y2, 'ro')
    plt.show()

sub_plots()
```

### Descriptive Plots

There are various ways you can modify your plots to make them more descriptive.

**X Axis:** plt.xlabel  
**Y Axis:** plt.ylabel  
**Title:** plt.title  
**Legend:** plt.legend  


```python
def descriptive_plots():
    t1 = np.arange(0.0, 5.0, 0.1)
    y1 = np.exp(-t1) * np.sin(2*np.pi*t1)
    t2 = np.arange(0.0, 5.0, 0.02)
    y2 = np.sin(2*np.pi*t2)

    plt.figure()
    plt.plot(t1, y1, 'b')
    plt.plot(t2, y2, 'r')
    
    plt.xlabel("Time")
    plt.ylabel("Size")
    plt.title("Descriptive Plot")
    plt.legend(('Graph 1', 'Graph 2'))
    plt.show()

descriptive_plots()
```
